```python
import folium
map=folium.Map(location=[19.6,85.8], zoom_start=6)
map.add_child(folium.Marker(location=[20.0,85.8],popup="Hi I am a Marker", icon=folium.Icon(color='green')))
map.save("Map2.html")
map
```

```python
import folium
map=folium.Map(location=[20.0,85.8], zoom_start=6)

##feature group
fg=folium.FeatureGroup(name="My Map")
#bbsr to jagannathpur
for coordinates in [[19.86,85.85],[20.2,85.84]]:
    fg.add_child(folium.Marker(location=coordinates,popup="Hi I am a Marker", icon=folium.Icon(color='green')))

map.add_child(fg)
map.save("Map2.html")
map
```
