Based on the above describe() function, we'll add/subtract '1.5xIQR' to/from 3rd/1st quartile of each data and can know whether there are outliers in the data set

```python
mcdIQR = mcddfdesc.iloc[6]-mcddfdesc.iloc[4]
mcdIQR*1.5
```

```python
mcddful=mcddfdesc.iloc[6]+(mcdIQR*1.5)
mcddful   #upperlimit
```

```python
mcddfll=mcddfdesc.iloc[4]-(mcdIQR*1.5)
mcddfll   #lowerlimit
```

```python
mcddfll<mcddfdesc.iloc[3]
```

```python
mcddful>mcddfdesc.iloc[7]
```

2. Thus we can conclude that where the upper limit value calculated is not greater than the max value for the column, there are outlier present in that column. Or where the lower limit value calculated is not smaller than the min value for the column, there are outlier present in that column

Plotting a correlation matrix of the dataset, we get the below table

```python
mcdcorr=mcddf.corr()
mcdcorr
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(mcdcorr, annot=True, vmin=-1, vmax=1)
```

Values > 0.8 and <=1 have been highlighted in light orange. So it can be concluded that 'Calories', 'Total Fat' & 'Saturated Fat' and their '% daily value' have very high positive correlation values.

Values >0.5 and <0.8 have been highlighted in medium orange. 
We can see that 'Carbohydrate', 'Cholesterol', 'Dietary Fiber', 'Sugars', 'Sodium', 'Iron' & 'Protien' have more than average positive correlation values

Also, we can unstack the corr matrix and find values >0.85 and <1, to single out those variables with highest correlation

```python
corruns = mcdcorr.unstack().sort_values(ascending=False)
corruns1 = corruns[corruns<1]
corruns2 = corruns1[corruns1>0.85]
corruns2
```

3. Above we can see the variables with highest correlation values are 'Calories', 'Calories from Fat','Total Fat' with values 0.904588,
